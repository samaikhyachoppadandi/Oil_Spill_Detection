```python
import cv2
import numpy as np
import glob
```

```python
IMAGES = "./oilspill/*"
```

```python
for filename in glob.glob(IMAGES):
    print(filename)
    image = cv2.imread(filename)
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    thresh, image_edges = cv2.threshold(image_gray, 100, 225, cv2.THRESH_BINARY)
    canvas = np.zeros(image.shape, np.uint8)
    canvas.fill(255)
    contours_draw, hierarchy = cv2.findContours(image_edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    img = cv2.drawContours(canvas, contours_draw, -1, (0, 0, 0), 3)
    x = filename.split("\\")
    y = x[1].split(".")
    new_file = y[0]+"_mask."+y[1]
    print(new_file)
    cv2.imwrite(new_file,img)
```

```python
image = cv2.imread("./oilspill/oil_spill_1.jpg")
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh, image_edges = cv2.threshold(image_gray, 100, 225, cv2.THRESH_BINARY)
canvas = np.zeros(image.shape, np.uint8)
canvas.fill(255)
contours_draw, hierarchy = cv2.findContours(image_edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
img = cv2.drawContours(canvas, contours_draw, -1, (0, 0, 0), 3)
cv2.imshow("thresh",thresh)
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
